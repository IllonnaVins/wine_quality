# Data Preparation

## Import Libraries

```python
import pandas as pd

from src.data.dataset import load_dataset, save_dataset

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
%load_ext autoreload
%autoreload 2
```

## Загрузка данных

```python
df = load_dataset()
```

## Заполнение пропущенных значений

Заменим пропуски в данных на среднее значение

```python
df = df.fillna(df.mean())

df.isna().sum()
```

## Категориальные данные

Заменим тип вина на 0 и 1

```python
le = LabelEncoder()
df.type = le.fit_transform(df.type)

df.head()
```

## Сборка финальных датасетов для моделирования

### Удаляем из датасета целевую переменную

```python
X = df.drop('quality', axis=1)
y = df.quality
```

### Нормализуем данные

```python
scaler = MinMaxScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

X.head()
```

```python
df_train = X
df_train['quality'] = y
```

```python
df_train.head()
```

```python
df_train.to_csv('train', encoding='utf-8', index=False)

save_dataset(df_train, 'train.csv')
```
